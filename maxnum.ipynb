```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

#max is equal to num 1
max = num1

#compare to max 2
if max < num2:
    max = num2

#compare to max 3
if max < num3:
    max = num3

print("Max number is", max)
```
